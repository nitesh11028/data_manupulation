```python
df_A1 = df[df.Age.between(40,50)]
df_A1
```

```python
df_A2 = df[df.Age.isin([40,50])] # only for the Age of 40,50
df_A2
```

```python
df_A3 = df[~df.Age.isin([40,50])] # Not the Age of 40,50
df_A3
```

```python
df_A4 = df[(df.Sex == 'female')&(df.Age <18)]
df_A4
```

```python
df_A4 = df[(df.Sex == 'female')&(df.Age <18)].count()
df_A4
```

```python
df['Length'] = df.Name.str.len()
df
```

```python
df['Gender-Male'] = df.Name.str.contains('Mr\.')
df
```

```python
df['Gemder-Female'] = df.Name.str.contains('Mrs\.','Miss\.')
df
```

```python
new = {'male':0,'female':1}
```

```python
df['new_gender'] = df.Sex.map(new)
df
```

```python
df[['Last-Name','First-name']] = df.Name.str.split(',',expand = True)
df
```

```python
df['Gender_la'] = df.Sex.apply(lambda x: 'M' if x == 'male'else 'F')
df
```

```python
df['Fare-Roundoff'] = df.Fare.apply(np.round)
df
```
